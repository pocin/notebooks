## Vybrat z komentaru ciselne hodnoty + procistit
zachyti vsechny cisla z komentu i ve formatu napr.:
urcite by se dalo vymyslet robustnejsi reseni, ale pro tyto ucely by to mohlo stacit


```python
def parse_comment(comment):
    result =  "".join(re.findall(r'\d+', comment, flags=re.IGNORECASE))
    return result if result else 0
```

```python
df = pd.DataFrame(comments['data'])

df['tip'] = df.message.apply(parse_comment)

df.tip = df.tip.astype(int)
```
